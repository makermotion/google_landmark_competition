```python
import numpy as np 
import pandas as pd 
import os
import seaborn as sns
from fastai import *
from fastai.vision import *
from fastai.metrics import error_rate
import re
import glob
from pathlib import Path
import random
import seaborn as sns
import matplotlib.pyplot as plt
random.seed(42)
import warnings
warnings.simplefilter("ignore", UserWarning)
```

```python
base_dir = Path('/home/scutum/WORK/downloads/')
```

```python
train = pd.read_csv(base_dir/'train.csv')
train["image_"] = 'train/'+train.id.str[0]+"/"+train.id.str[1]+"/"+train.id.str[2]+"/"+train.id+".jpg"
train['impath'] = [Path(i) for i in train['image_']]
train['fnames'] = train.id+".jpg"
train["target_"] = train.landmark_id.astype(str)
```

```python
most_freq_1000 = train['landmark_id'].value_counts().iloc[:1000].index.tolist()
```

```python
train = train[train['landmark_id'].isin(most_freq_1000)]
```

```python
data = ImageDataBunch.from_df('/home/scutum/WORK/downloads/', df=train , valid_pct=.07, fn_col='image_', label_col='landmark_id', size=(128,128), num_workers=16, bs=128)
```

```python
data.show_batch(rows=3, figsize=(8,7))
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(3)
```

```python
learn.export('/home/scutum/WORK/scutum/notebooks/glr.pkl')
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(1)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.export('/home/scutum/WORK/scutum/notebooks/glr.pkl')
```
